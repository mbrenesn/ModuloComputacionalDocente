En un lenguaje interpretado:
* Un programa llamado *interpretador* toma las instrucciones y las traduce a lenguaje de máquina.
* No existe etapa de compilación.
* A diferencia de un compilador, el cual toma todas las instrucciones y genera un ejecutable en lenguaje de máquina, un interpretador toma el código **línea por línea** y las ejecuta.

Algunas razones por las cuales es importante el uso de la programación interpretada en computación científica:
* Facilidad en la curva de aprendizaje.
* El número de bibliotecas disponibles es mucho más grande, en general.
* Excelente soporte para aplicaciones científicas.
* Ciclos de desarrollo más rápidos.

Python es un lenguaje multipropósito y multiparadigma, i.e., el lenguaje no enforza el uso de paradigmas.
* La sintaxis es *usualmente* limpia, pero puede ser muy difícil de optimizar.
* Altamente extendible.

### *Type system* (sistema de interpretación)

````python
# Strong typing

'foo' + 5 # Error al runtime
````

````python
# Dynamic typing

a = 'foo'
b = 2 * a # b = 'foofoo'
print(b)
a = 5
b = 2 * a # b = 10
print(b)
````

````python
# 'Duck' typing
# “If it walks like a duck and quacks like a duck, then it’s a duck.”
# Esto quiere decir que Python intenta realizar las operaciones sin confirmar 
# si los objetos se comportan de la forma esperada, igual realiza la operación.
# E.g.:
def foo(a, b):
    return a + b

a = 5
b = 7
print(foo(a, b))
````

Básicamente, en `Python`, el interpretador no se pregunta `¿Qué cosa es este objeto?`. En lugar de esto, se pregunta `¿Este objeto puede realizar esta operación?`

````python
# Pero si, por ejemplo, los objetos no implementan la operación, el sistema igual intenta realizarla
# y devuelve un runtime error
a = 'foo'
b = 2
print(foo(a, b))
````

### Flujos de control

Los `for loops` son flujos iterativos en los cuales **conocemos la cantidades de iteraciones** a realizar.
```python
    for i in list_i:
        baz(i)
```
Los `while loops` son comunes para casos en los cuales desconocemos la cantidad de iteraciones (e.g., iterar hasta alcanzar cierta tolerancia).
```python
    while a > b:
        foo()
        bar()
```
Los casos aislados se pueden controlar usando flujos `if-else`, cuya sintaxis es:
```python
    if a > b:
        foo()
    elif b != c:
        bar()
    else:
        baz()
```

Las siguientes palabras clave son importantes para flujo de control:
```python
    pass
    break
    continue
```
* `pass` se usa para literalmente no hacer nada, o como `placeholder`.
* `break` se utiliza para cortar un proceso iterativo (e.g., parar de iterar porque se alcanzó la condición de convergencia).
* `continue` se utiliza para saltarse la iteración actual y seguir con la que continúa.

### Funciones

Las funciones son de las construcciones más fundamentales de un programa. Dependiendo de su diseño, permiten manipular un `estado` o realizar operaciones autocontenidas

````python
# Sintaxis
# Funciones con valor de retorno
def func(a, b, c):
    a = 3 * b
    return a + b + c
# Funciones sin valor de retorno
def func2(a , b , c):
    a = 3 * b
    c = 2 * a
````

Las funciones que tienen valor de retorno se invocan de distinta forma a las funciones que no tienen valor de retorno. Por ejemplo, las funciones anteriores se invocan de la siguiente forma:
```python
d = func(aVal, bVal, cVal)
```
```python
func2(aVal2, bVal2, cVal2)
```

El manejo de excepciones suele ser una forma robusta de predecir errores que pueden ocurrir ya sea a nivel de interpretación o al nivel de ejecución. Por ejemplo,

### Formatos de impresión

Existen dos formas de realizar distintos formatos de variables y caracteres.

La primera es mediante el operador `%`:

````python
print("Hoy comí %d %s para el desayuno" % (2, "manzanas"))
````

### Arreglos y colecciones de datos

En `Python`, no existe el concepto de el uso de referencias de memoria para construir arreglos de datos.
En lugar de esto, `Python` introduce estructuras `mutables` e `inmutables`. 

````python
# Esto es una lista.
# Las listas son estructuras de datos primitivas en Python que pueden representar, e.g., vectores, matrices y tensores.
# Las listas son *mutables*, i.e., una función puede modificar su contenido directamente en memoria:
a = [3, 1, 'foo', 12.0]
def func(l):
    l[0] = 5
    return l

print(func(a))
````

````python
# Esto es una tupla.
# Las tuplas son muy similares a las listas, solo que son *inmutables*:
b = (3, 1, 'foo', 12.0)

print(func(b))
````

En `Python`, el concepto de mutabilidad se utiliza para manipular los recursos de la máquina de una forma abstracta. De forma sucinta, la mutabilidad resuelve el problema de si hacer o no hacer copias locales de estructuras de datos dependiendo del objetivo deseado (*pasar por valor vs pasar por referencia*)

Los diccionarios son estructuras de datos múy útiles en ciertas aplicaciones, dado que dan lugar a algoritmos de búsqueda muy eficientes a cambio de incrementar la cantidad de recursos requerida para su implementación.
Los diccionarios son mutables:

````python
d = {'name': 'Marlon', 'age': 21}
print(d['name'])
print(d['age'])
````

````python
def func(dd):
    dd['age'] = 20
    return dd

print(func(d))
````

### `numpy.ndarray`

El `array` es el componente fundamental de `NumPy`. 
* Diseñado para accesis igual a como se hace con una lista de `Python`
* Todos sus elementos son del mismo tipo
* Idealmente diseñados para almacenar y manipular muchos elementos 

````python
# Constructor
a = np.array([1, 4, 5 , 8], dtype = float)

print(a)
print(type(a))
````

````python
# La siguiente operación se llama "array slicing"
print(a[:2])
print(a[3])
````

Los `numpy.ndarray` también pueden ser de múltiples dimensiones como las listas:

````python
a = np.array([[1, 2, 3], [4, 5, 6]])

print("Matriz a:\n", a)
# Una dereferencia de una matriz en una sola coordenada devuelve una fila:
print("Primera fila de a:", a[0])
# En ambas coordenadas, se accesa el elemento:
print("Elemento (1,2) de a:", a[0, 1])
# Los numpy arrays son objetos que tienen comportamiento y atributos.
# Los métodos son funciones que se accesan mediante el operador "." y los atributos 
# se accesan con el mismo operador, pero sin los paréntesis indicando función: 
print("Dimensión de a:", a.shape)
print("Promedio de los elementos de a:", a.mean())
````

La forma y dimensiones de los arreglos se pueden modificar:

````python
a = np.array(range(10))
print(a)
````

````python
b = a.reshape((5, 2))
print(b)
````

Las copias deben ser explícitas, de lo contrario, en algunas situaciones `NumPy` crea punteros en memoria de arreglos ya construidos (más de esto también más adelante):

````python
a = np.array([1, 2, 3])
b = a # b es un "shallow copy" de a: modificar a modifica b
c = a.copy() # c es un "deep copy" de a: modificar a no modifica c, se duplica en memoria

# Veamos:
print("Antes de modificar a:")
print("a:", a)
print("b:", b)
print("c:", c)

a[0] = 0

print("Después de modificar a:")
print("a:", a)
print("b:", b)
print("c:", c)
````

Las operaciones están sobrecargadas, de manera tal que realizar operaciones aritméticas aplica la operación **elemento por elemento**:

````python
a = np.array([1, 2, 3], dtype = float)
b = np.array([5, 2, 6], dtype = float)

print(a + b)
print(a - c)
print(a * b)
print(b ** a)
````

`NumPy` ofrece una biblioteca extensa de funciones matemáticas comunes que pueden ser aplicadas de manera **vectorial**: `np.abs`, `np.sign`, `np.sqrt`, `np.log`, `np.sin`, `np.arctan`...

````python
a = np.linspace(0.3, 0.6, 4)
print(a)
print(np.sin(a))
````

Incluso funciones para análisis estadístico:

````python
a = np.array([2, 1, 9], dtype = float)
print(a.mean())
print(a.var())
print(a.std())
````

De las funcionalidades más importantes de `NumPy` tenemos las **operaciones fundamentales del álgebra lineal**. Dichas subrutinas son precompiladas y enlazadas con la biblioteca, de manetal que se puede esperar eficiencia similar a las rutinas usandas en `C/C++`:

````python
a = np.array([[0.0, 1.0], [2.0, 3.0]])
b = np.array([2.0, 3.0])
c = np.array([[1.0, 1.0], [4.0, 0.0]])

print(np.dot(a, b))
print(np.dot(b, a))
print(np.dot(a, c))
print(np.dot(c, a))
````

Algunas rutinas se encuentran dentro del módulo `numpy.linalg`:

````python
a = np.array([[4, 2, 0], [9, 3, 7], [1, 2, 1]], dtype = float)
print(a)
````

````python
print(np.linalg.det(a))
````

````python
# Esto usa un algoritmo QR en su "backend", más de esto más adelante
eigvals, eigvecs = np.linalg.eig(a)
print(eigvals)
print(eigvecs)
````

## Práctica guíada

Es sencillo demostrar por sustitución trigonométrica que
$$
\pi = \int_{0}^{1} {\textrm{d}}x\;\frac{4}{1 + x^2}.
$$
Estudiaremos de manera más formal a nivel matemático como podemos implementar algoritmos computacionales para evaluar integrales. De momento, consideremos el caso sencillo de aproximar la integral mediante **una suma de Riemann de punto medio**:
$$
\pi \approx \sum_{n=1}^{N} h * f(x_n),
$$
donde $h \equiv (b - a) / N$, $b$ y $a$ son los límites superior e inferior de la integral; respectivamente, y $N$ es la cantidad de rectángulos de la suma de Riemann. A su vez, $f(x) = 4 / (1 + x^2)$ y $x_n$ son los llamados **puntos de colocación**, que en este caso, corresponden a los valores medios para evaluar la altura de los rectángulos.
